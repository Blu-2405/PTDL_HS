```
from nltk.tokenize import word_tokenize
text = "This is a simple example to demonstrate stopword removal."
tokens = word_tokenize(text)
filtered = [w for w in tokens if w.lower() not in stopwords.words('english')]
print(filtered)
```

```
from nltk.corpus import wordnet
word = "computer"
for s in wordnet.synsets(word)[:3]:
    print("Định nghĩa:", s.definition())
    print("Ví dụ:", s.examples())
```

7

```
synonyms, antonyms = [], []
for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())
print("Đồng nghĩa:", set(synonyms))
print("Trái nghĩa:", set(antonyms))
```

```
from nltk.corpus import names
male_names = names.words('male.txt')
female_names = names.words('female.txt')

print("Số tên nam:", len(male_names))
print("Số tên nữ:", len(female_names))
print("10 tên nam đầu:", male_names[:10])
print("10 tên nữ đầu:", female_names[:10])
```

12

```
import random
print("15 tên nam:", random.sample(male_names, 15))
print("15 tên nữ:", random.sample(female_names, 15))
```

  13

```
labeled_names = ([(name, 'male') for name in male_names] +
                 [(name, 'female') for name in female_names])
features = [(name[-1], gender) for (name, gender) in labeled_names]
print(features[:10])
```
